# Question1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import(NeighborhoodComponentsAnalysis,KNeighborsClassifier)
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
```

```python
X={'Outlook':[0,0,1,2,2,2,1,0,0,2],
      'Temperature':[0,0,0,1,2,2,2,1,2,1],
      'Humidity':[1,1,1,1,0,0,0,1,0,0] ,
      'Wind':[0,1,0,0,0,1,1,0,0,0] ,
     }
Y={'PlayTennis':[0,0,1,1,1,0,1,0,1,1]}
Xx=pd.DataFrame(X)
Yy=pd.DataFrame(Y)
```

```python
X_train=Xx[['Outlook','Temperature','Humidity', 'Wind']]
Y_train=Yy['PlayTennis']
X_train
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, Y_train)
```

```python
test_x={'Outlook':[0,1,1,2],'Temperature':[1,1,0,1],'Humidity':[0,1,0,1],'Wind':[1,1,0,1]}
test_y={'PlayTennis':[1,1,1,0]}
```

```python
Xx=pd.DataFrame(test_x)
Yy=pd.DataFrame(test_y)
```

```python
y_predict=knn.predict(Xx)
print("Predict : ",y_predict)
```

```python
Accuracy1=accuracy_score(Yy,y_predict)
Accuracy1
```

# Question 2

```python
csv_data=pd.read_csv("cancer.csv")
```

```python
csv_data['Diagnosis']=csv_data['Diagnosis'].map({'B':0,'M':1})

Y=csv_data['Diagnosis']
Y=pd.DataFrame(Y)
```

```python
X=csv_data
```

```python
X=X.drop('Diagnosis',axis=1)
X=X.drop('Id',axis=1)
```

```python
X
```

```python
X_array=np.array(X)
X_array
```

```python

Y
```

```python
Y_array=np.array(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_array, Y_array,stratify=Y, test_size=0.2, random_state=42)
```

```python
nca = NeighborhoodComponentsAnalysis(random_state=42)
```

```python
nca.fit(X_train, y_train)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_predict=knn.predict(X_test)
print("Predict : ",y_predict)
```

```python
Accuracy1=accuracy_score(y_test,y_predict)
print("Accuracy : ",Accuracy1)
```
