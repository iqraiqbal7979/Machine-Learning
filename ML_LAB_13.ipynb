```python
import pandas as pd
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
train_data = pd.DataFrame({
    'X1': [60, 62, 67, 70, 71, 72, 75, 78],
    'X2': [22, 25, 24, 20, 15, 14, 14 , 11],
    'X3': [15, 15, 14, 14, 13, 12 , 12 , 11],
    'X4': [2.5 ,2.7,2.9,2.8,3.5,3.8,3.9,4.1],
    'X5': [6, 7.5 ,9, 11, 14, 15.5, 18, 21],
    'Y': [140, 155, 159, 179, 192, 200, 212 , 215]
})
```

```python
X = train_data[['X1', 'X2', 'X3', 'X4', 'X5']]
y = train_data['Y']

regul = [0.01, 0.1, 1]
result = {}
```

```python
for alpha in regul:
   
    model = Lasso(alpha=alpha)
    
   
    model.fit(X, y)
    
   
    y_pred = model.predict(X)
    
    
    regression_score = model.score(X, y)
    
    
    error = mean_squared_error(y, y_pred)
    
    
    result[alpha] = {'Regression Score': regression_score, 'Error': error}
```

```python
result_df = pd.DataFrame(result).T 

print(result_df)
```

```python
# Import necessary libraries
import pandas as pd
from sklearn import linear_model
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Read the dataset
df1 = pd.read_csv('Real Estate.csv')

# Define the features and target
X = df1.drop(columns=['Y house price of unit area'])  
Y = df1['Y house price of unit area'] 

# Set the regularization parameter
alpha = 0.01

# Lasso Regression
clf = linear_model.Lasso(alpha=alpha)
clf.fit(X, Y)
p_lasso = clf.predict(X)
lasso_score = r2_score(Y, p_lasso)
lasso_error = np.sqrt(mean_squared_error(Y, p_lasso))

# Ridge Regression
ridge = Ridge(alpha=alpha)
ridge.fit(X, Y)
p_ridge = ridge.predict(X)
ridge_score = r2_score(Y, p_ridge)
ridge_error = np.sqrt(mean_squared_error(Y, p_ridge))

# Print the results
print("Lasso Regularization:")
print(f"R2 Score: {lasso_score}")
print(f"RMSE: {lasso_error}")

print("\nRidge Regularization:")
print(f"R2 Score: {ridge_score}")
print(f"RMSE: {ridge_error}")
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error, r2_score

# Creating the dataset
data = {
    'X1': [60, 62, 67, 70, 71, 72, 75, 78],
    'X2': [22, 25, 24, 20, 15, 14, 14, 11],
    'X3': [15, 15, 14, 14, 13, 12, 12, 11],
    'X4': [2.5, 2.7, 2.9, 2.8, 3.5, 3.8, 3.9, 4.1],
    'X5': [6, 7.5, 9, 11, 14, 15.5, 18, 21],
    'Y': [140, 155, 159, 179, 192, 200, 212, 215]
}

# Convert the data into a Pandas DataFrame
df = pd.DataFrame(data)

# Define the input variables (X) and the target variable (Y)
X = df[['X1', 'X2', 'X3', 'X4', 'X5']].values
y = df['Y'].values

# List of regularization parameters to test
alphas = [0.01, 0.1, 1]

# Initialize dictionaries to store results
lasso_results = {'Regularization Parameter': [], 'Regression Score': [], 'Error': []}
ridge_results = {'Regularization Parameter': [], 'Regression Score': [], 'Error': []}

# Perform Lasso Regularization
for alpha in alphas:
    lasso = Lasso(alpha=alpha)
    lasso.fit(X, y)
    y_pred_lasso = lasso.predict(X)
    lasso_results['Regularization Parameter'].append(alpha)
    lasso_results['Regression Score'].append(r2_score(y, y_pred_lasso))
    lasso_results['Error'].append(mean_squared_error(y, y_pred_lasso))

# Perform Ridge Regularization
for alpha in alphas:
    ridge = Ridge(alpha=alpha)
    ridge.fit(X, y)
    y_pred_ridge = ridge.predict(X)
    ridge_results['Regularization Parameter'].append(alpha)
    ridge_results['Regression Score'].append(r2_score(y, y_pred_ridge))
    ridge_results['Error'].append(mean_squared_error(y, y_pred_ridge))

# Convert the results to DataFrames
lasso_df = pd.DataFrame(lasso_results)
ridge_df = pd.DataFrame(ridge_results)

# Print the results
print("Lasso Regularization Results:")
print(lasso_df)
print("\nRidge Regularization Results:")
print(ridge_df)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Step 1: Read the dataset from the CSV file
df1 = pd.read_csv('Real Estate.csv')  # Replace with the actual path of your CSV file

# Step 2: Prepare the data (Input features and target)
X = df1.drop(columns=['Y house price of unit area'])  # Input features
Y = df1['Y house price of unit area']  # Target variable

# Step 3: Split the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Regularization parameters to test
alphas = [0.01, 0.1, 1]

# Initialize dictionaries to store results for Lasso and Ridge
lasso_results = {'Regularization Parameter': [], 'R2 Score': [], 'RMSE': []}
ridge_results = {'Regularization Parameter': [], 'R2 Score': [], 'RMSE': []}

# Step 4: Perform Lasso Regularization
for alpha in alphas:
    lasso = Lasso(alpha=alpha)
    lasso.fit(X_train, y_train)
    y_pred_lasso = lasso.predict(X_test)
    lasso_results['Regularization Parameter'].append(alpha)
    lasso_results['R2 Score'].append(r2_score(y_test, y_pred_lasso))
    lasso_results['RMSE'].append(np.sqrt(mean_squared_error(y_test, y_pred_lasso)))

# Step 5: Perform Ridge Regularization
for alpha in alphas:
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train, y_train)
    y_pred_ridge = ridge.predict(X_test)
    ridge_results['Regularization Parameter'].append(alpha)
    ridge_results['R2 Score'].append(r2_score(y_test, y_pred_ridge))
    ridge_results['RMSE'].append(np.sqrt(mean_squared_error(y_test, y_pred_ridge)))

# Step 6: Convert results to DataFrames for easy viewing
lasso_df = pd.DataFrame(lasso_results)
ridge_df = pd.DataFrame(ridge_results)

# Step 7: Display the results
print("Lasso Regularization Results:")
print(lasso_df)
print("\nRidge Regularization Results:")
print(ridge_df)

# Step 8: Compare the results and make an observation
```
