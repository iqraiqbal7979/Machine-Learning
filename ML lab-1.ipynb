```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = {'X1': [1, 3.5, 0.75, 2.5, 1.5], 'X2': [2, 1.5, 1.25, 2, 0.5], 'Y': [2, 2.5, 1.5, 3.25, 1]}
df = pd.DataFrame(data)

def hypothesis(x1_values, x2_values, theta0, theta1, theta2):
    predictions = []
    for x1, x2 in zip(x1_values, x2_values):
        hypo = theta0 + theta1 * x1 + theta2 * x2
        predictions.append(hypo)
    return predictions


def mae(predictions, targets):
    sample_num = len(predictions)
    sample_error = 0.0
    for pred, targ in zip(predictions, targets):
        sample_error += np.abs(pred - targ)
    mae_error = (1.0 / sample_num) * sample_error
    return mae_error


def mse(predictions, targets):
    sample_num = len(predictions)
    sample_error = 0.0
    for pred, targ in zip(predictions, targets):
        sample_error += (pred - targ) ** 2
    mse_error = (1.0 / (2 * sample_num)) * sample_error
    return mse_error


def gradient_descent(iterations, x1_values, x2_values, y_values, theta0_init, theta1_init, theta2_init):
    theta0 = theta0_init
    theta1 = theta1_init
    theta2 = theta2_init
    alpha = 0.05

    for i in range(iterations):
        predictions = hypothesis(x1_values, x2_values, theta0, theta1, theta2)
        errors = [pred - actual for pred, actual in zip(predictions, y_values)]
        
        theta0 -= (alpha / len(x1_values)) * sum(errors)
        theta1 -= (alpha / len(x1_values)) * sum(error * x1 for error, x1 in zip(errors, x1_values))
        theta2 -= (alpha / len(x1_values)) * sum(error * x2 for error, x2 in zip(errors, x2_values))
        
        if i % 10 == 0: 
            print(f"Iteration {i}: θ0={theta0:.4f}, θ1={theta1:.4f}, θ2={theta2:.4f}")
    
    return theta0, theta1, theta2


theta0_init = 0.5
theta1_init = 0.4
theta2_init = 0.9


optimized_theta0, optimized_theta1, optimized_theta2 = gradient_descent(50, df['X1'], df['X2'], df['Y'], theta0_init, theta1_init, theta2_init)
print(f"Optimized θ0: {optimized_theta0}")
print(f"Optimized θ1: {optimized_theta1}")
print(f"Optimized θ2: {optimized_theta2}")


predictions = hypothesis(df['X1'], df['X2'], optimized_theta0, optimized_theta1, optimized_theta2)
print(f"Predictions: {predictions}")


plt.scatter(df['X1'], df['Y'], color="blue", label="Data Points")
plt.plot(df['X1'], predictions, color="red", linestyle='-', label="Linear Regression")
plt.xlabel("X1")
plt.ylabel("Y")
plt.title("X1 vs Y Graph")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = {'X1': [1, 3.5, 0.75, 2.5, 1.5], 'X2': [2, 1.5, 1.25, 2, 0.5], 'Y': [2, 2.5, 1.5, 3.25, 1]}
df = pd.DataFrame(data)


def hypothesis(x1, x2, theta0, theta1, theta2):
    return theta0 + theta1 * x1 + theta2 * x2

# Mean Absolute Error (MAE) Function
def calculate_mae(predictions, targets):
    sample_size = len(predictions)
    error_sum = np.sum(np.abs(predictions - targets))
    return error_sum / sample_size

# Mean Squared Error (MSE) Function
def calculate_mse(predictions, targets):
    sample_size = len(predictions)
    error_sum = np.sum((predictions - targets) ** 2)
    return error_sum / (2 * sample_size)

# Gradient Descent Function
def gradient_descent(iterations, x1_values, x2_values, y_values, theta0_init, theta1_init, theta2_init):
    theta0, theta1, theta2 = theta0_init, theta1_init, theta2_init
    alpha = 0.05
    
    for i in range(iterations):
        predictions = hypothesis(x1_values, x2_values, theta0, theta1, theta2)
        errors = predictions - y_values
        
        theta0 -= alpha * np.sum(errors) / len(x1_values)
        theta1 -= alpha * np.sum(errors * x1_values) / len(x1_values)
        theta2 -= alpha * np.sum(errors * x2_values) / len(x1_values)
        
        if i % 10 == 0:
            print(f"Iteration {i}: θ0={theta0:.4f}, θ1={theta1:.4f}, θ2={theta2:.4f}")
    
    return theta0, theta1, theta2

# Initial Parameters
theta0_init = 0.5
theta1_init = 0.4
theta2_init = 0.9

# Perform Gradient Descent
optimized_theta0, optimized_theta1, optimized_theta2 = gradient_descent(100, df['X1'], df['X2'], df['Y'], theta0_init, theta1_init, theta2_init)
print(f"Optimized θ0: {optimized_theta0}")
print(f"Optimized θ1: {optimized_theta1}")
print(f"Optimized θ2: {optimized_theta2}")

# Final Predictions
predictions = hypothesis(df['X1'], df['X2'], optimized_theta0, optimized_theta1, optimized_theta2)
print(f"Predictions: {predictions}")

# Plotting
plt.scatter(df['X1'], df['Y'], color="blue", label="Data Points")
plt.plot(df['X1'], predictions, color="red", linestyle='-', label="Linear Regression")
plt.xlabel("X1")
plt.ylabel("Y")
plt.title("X1 vs Y Graph")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Linear regression implementation from scratch
def calculate_coefficients(X, y):
    """
    Calculate the coefficients (slope and intercept) for linear regression.
    Formula for slope (b1) and intercept (b0):
    b1 = Σ((Xi - mean(X)) * (Yi - mean(y))) / Σ((Xi - mean(X))^2)
    b0 = mean(y) - b1 * mean(X)
    """
    # Mean of X and y
    X_mean = np.mean(X)
    y_mean = np.mean(y)

    # Calculate b1 (slope)
    numerator = 0
    denominator = 0
    for i in range(len(X)):
        numerator += (X[i] - X_mean) * (y[i] - y_mean)
        denominator += (X[i] - X_mean) ** 2
    b1 = numerator / denominator

    # Calculate b0 (intercept)
    b0 = y_mean - b1 * X_mean

    return b0, b1

def predict(X, b0, b1):
    """
    Use the coefficients to predict the target value.
    Formula: y = b0 + b1 * X
    """
    return b0 + b1 * X

def mean_squared_error(y_true, y_pred):
    """
    Calculate Mean Squared Error (MSE) between actual and predicted values.
    """
    return np.mean((y_true - y_pred) ** 2)

def r_squared(y_true, y_pred):
    """
    Calculate R-squared to evaluate the performance of the model.
    Formula: 1 - (Σ(y_true - y_pred)^2 / Σ(y_true - mean(y_true))^2)
    """
    ss_total = np.sum((y_true - np.mean(y_true)) ** 2)
    ss_residual = np.sum((y_true - y_pred) ** 2)
    return 1 - (ss_residual / ss_total)

# Generate some example data (you can use your own dataset)
# np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Flatten X and y to 1D arrays
X = X.flatten()
y = y.flatten()

# Calculate coefficients
b0, b1 = calculate_coefficients(X, y)

# Make predictions
y_pred = predict(X, b0, b1)

# Evaluate the model
mse = mean_squared_error(y, y_pred)
r2 = r_squared(y, y_pred)

print(f"Intercept (b0): {b0}")
print(f"Coefficient (b1): {b1}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R²): {r2}")

# Plot the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('X (Feature)')
plt.ylabel('y (Target)')
plt.legend()
plt.title('Linear Regression from Scratch')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample Data: Hours Studied vs. Test Score
X = np.array([1, 2, 3, 4, 5])
y = np.array([55, 60, 65, 70, 75])

theta0 = 0.0  # Intercept
theta1 = 0.0  # Slope

def predict(X, theta0, theta1):
    return theta0 + theta1 * X

def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def compute_gradients(X, y, y_pred):
    n = len(X)
    d_theta0 = (-2 / n) * np.sum(y - y_pred)
    d_theta1 = (-2 / n) * np.sum((y - y_pred) * X)
    return d_theta0, d_theta1

def update_parameters(theta0, theta1, d_theta0, d_theta1, learning_rate):
    theta0 = theta0 - learning_rate * d_theta0
    theta1 = theta1 - learning_rate * d_theta1
    return theta0, theta1

def gradient_descent(X, y, theta0, theta1, learning_rate, iterations):
    mse_history = []
    
    for i in range(iterations):
        y_pred = predict(X, theta0, theta1)
        mse = mean_squared_error(y, y_pred)
        mse_history.append(mse)
        d_theta0, d_theta1 = compute_gradients(X, y, y_pred)
        theta0, theta1 = update_parameters(theta0, theta1, d_theta0, d_theta1, learning_rate)
        
        if (i + 1) % 100 == 0:
            print(f"Iteration {i + 1}: MSE={mse:.4f}, theta0={theta0:.4f}, theta1={theta1:.4f}")
    
    return theta0, theta1, mse_history

learning_rate = 0.01
iterations = 100

theta0, theta1, mse_history = gradient_descent(X, y, theta0, theta1, learning_rate, iterations)

print(f"\nTrained Parameters:\nIntercept (theta0): {theta0}\nSlope (theta1): {theta1}")

y_pred = predict(X, theta0, theta1)

final_mse = mean_squared_error(y, y_pred)
final_r2 = r_squared(y, y_pred)

print(f"\nFinal Mean Squared Error (MSE): {final_mse}")
print(f"Final R-squared (R²): {final_r2}")

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(mse_history, color='blue')
plt.xlabel('Iterations')
plt.ylabel('Mean Squared Error')
plt.title('Cost Function Convergence')

plt.subplot(1, 2, 2)
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Hours Studied (X)')
plt.ylabel('Test Score (y)')
plt.legend()
plt.title('Linear Regression Fit')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample Data: Hours Studied vs. Test Score
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshaping to 2D array
y = np.array([55, 60, 65, 70, 75])  # Dependent variable (Test Score)

# Create the linear regression model
model = LinearRegression()

# Fit the model to the data (train the model)
model.fit(X, y)

# Make predictions using the trained model
y_pred = model.predict(X)

# Get the parameters (intercept and slope) of the model
theta0 = model.intercept_  # Intercept (the value of y when X is 0)
theta1 = model.coef_[0]  # Slope (the change in y for a unit change in X)

# Print the parameters
print(f"Intercept (theta0): {theta0}")  # Display the intercept
print(f"Slope (theta1): {theta1}")  # Display the slope

# Plot the actual data points
plt.scatter(X, y, color='blue', label='Actual Data')  # Scatter plot of actual data points

# Plot the regression line based on the predictions
plt.plot(X, y_pred, color='red', label='Regression Line')  # Line showing predicted values
plt.xlabel('Hours Studied (X)')  # Label for the x-axis
plt.ylabel('Test Score (y)')  # Label for the y-axis
plt.legend()  # Show the legend
plt.title('Linear Regression Fit')  # Title for the plot
plt.show()  # Display the plot
```
