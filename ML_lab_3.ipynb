```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame from a dictionary
data = {'X1': [60, 62, 67, 70, 71, 72, 75, 78],
        'X2': [22, 25, 24, 20, 15, 14, 14, 11],
        'Y': [0, 0, 0, 0, 1, 1, 1, 1]}
data = pd.DataFrame(data)

# Initial parameters
theta0 = 0.21
theta1 = 0.3
theta2 = 0.033
Y = data['Y']

# Sigmoid function
def sigmoid(z):
    return (1 / (1 + np.exp(-z)))

# Hypothesis function
def hypothesis(data, theta0, theta1, theta2):
    h = theta0 + theta1 * data['X1'] + theta2 * data['X2']
    return sigmoid(h)

# Cost function
def cost_function(Y, hypo):
    m = len(Y)
    return -1 / m * np.sum(Y * np.log(hypo) + (1 - Y) * np.log(1 - hypo))

# Gradient descent function
def gradient_descent(Y, X1, X2, iterations, alpha):
    m = len(Y)
    theta0 = np.random.rand()
    theta1 = np.random.rand()
    theta2 = np.random.rand()
    cost_history = []

    for i in range(iterations):
        hypo = hypothesis(data, theta0, theta1, theta2)

        # Calculate gradients
        theta0_gradient = np.sum(hypo - Y) / m
        theta1_gradient = np.sum((hypo - Y) * X1) / m
        theta2_gradient = np.sum((hypo - Y) * X2) / m

        # Update parameters
        theta0 -= alpha * theta0_gradient
        theta1 -= alpha * theta1_gradient
        theta2 -= alpha * theta2_gradient
        
        # Save cost value
        cost = cost_function(Y, hypo)
        cost_history.append(cost)

    return theta0, theta1, theta2, cost_history

# Setting hyperparameters and running gradient descent
iterations = 100
alpha = 0.02
X1 = data['X1']
X2 = data['X2']
theta0, theta1, theta2, cost_history = gradient_descent(Y, X1, X2, iterations, alpha)

print("Final Parameters:", theta0, theta1, theta2)
print("\n\nCost History:\n", cost_history[-10:])


```

```python
# Plotting the S-curve (Sigmoid Function)
plt.figure(figsize=(6, 5))
# Creating a range of values for X1
x1_values = np.linspace(data['X1'].min(), data['X1'].max(), 100)
x2_fixed = data['X2'].mean()  # Fix X2 at its mean value
# Calculate the predicted probabilities for each X1 value
z_values = theta0 + theta1 * x1_values + theta2 * x2_fixed
probabilities = sigmoid(z_values)
# Plotting
plt.plot(x1_values, probabilities, label='S-Curve (Predicted Probability)', color='blue')
plt.scatter(data['X1'], Y, color='red', label='Data Points')
# plt.scatter(data['X2'], Y, color='purple', label='Data Points')
plt.xlabel('X1')
plt.ylabel('Predicted Probability')
plt.title('Logistic Regression S-Curve')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Creating a DataFrame from a dictionary
data = {
    'X1': [60, 62, 67, 70, 71, 72, 75, 78],
    'X2': [22, 25, 24, 20, 15, 14, 14, 11],
    'Y': [0, 0, 0, 0, 1, 1, 1, 1]
}
data = pd.DataFrame(data)

# Features and target variable
X = data[['X1', 'X2']]  # Feature matrix
y = data['Y']  # Target variable

# Creating and fitting the model
model = LogisticRegression()
model.fit(X, y)

# Predicting probabilities
data['Predicted Probability'] = model.predict_proba(X)[:, 1]

# Plotting the S-Curve (Sigmoid Function)
plt.figure(figsize=(6, 5))

# Creating a range of values for X1
x1_values = np.linspace(data['X1'].min(), data['X1'].max(), 100)
x2_fixed = data['X2'].mean()  # Fix X2 at its mean value
X_new = np.column_stack((x1_values, np.full_like(x1_values, x2_fixed)))

# Calculate the predicted probabilities for each X1 value
probabilities = model.predict_proba(X_new)[:, 1]

# Plotting
plt.plot(x1_values, probabilities, label='S-Curve (Predicted Probability)', color='blue')
plt.scatter(data['X1'], data['Y'], color='red', label='Data Points')
plt.xlabel('X1')
plt.ylabel('Predicted Probability')
plt.title('Logistic Regression S-Curve')
# plt.axhline(0.5, color='gray', linestyle='--')  # Decision boundary
# plt.axvline(0, color='gray', linestyle='--')  # Vertical line at 0
plt.legend()
plt.grid()
plt.show()
```
